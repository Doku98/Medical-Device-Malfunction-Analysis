```python
# given one of the dataframes in the cell below and a column name, returns the amount of times specific values in that dataframe occurs in that column
# e.g.count_occurance_in_column_per_year(data_1999,"event_type") will return how many times different event_types occur in the year 1999
def count_occurance_in_column(data, column):
    data_by_event_type_count = data.groupby(column).count()
    data_by_event_type_count = data_by_event_type_count.assign(Count =  data_by_event_type_count.get("exemptn_no")).get(["Count"])
    return data_by_event_type_count
```

```python
#The cell below shows how A/B testing (If the different between needels have the same rate of malfuctions due to random chance or some other factor)
#Null Hypothesis: The probability of a malfuncting needle being stright is the same as the probability of a malfuncting needle being curved 
#Alternative Hypothesis: The probability of a malfuncting needle being stright is less than the probability of a malfuncting needle being curved 


curved_or_stright_needles = data_1999_2019_needles[(data_1999_2019_needles.get("brand_name") == "CURVED NEEDLE DELIVERY SYSTEM") | (data_1999_2019_needles.get("brand_name") == "STRAIGHT NEEDLE DELIVERY SYSTEM") ]

total_needles = curved_or_stright_needles.shape[0]

straight_needles_malfunction = data_1999_2019_needles[data_1999_2019_needles.get("brand_name") == "STRAIGHT NEEDLE DELIVERY SYSTEM"]
curved_needles_malfunction = data_1999_2019_needles[data_1999_2019_needles.get("brand_name") == "CURVED NEEDLE DELIVERY SYSTEM"]

proportion_straight_malfunction = straight_needles_malfunction.shape[0]/total_needles
proportion_curved_malfunction = curved_needles_malfunction.shape[0]/total_needles


print("Straight of Curved needle malfunctions in our data")
print(straight_needles_malfunction.shape[0])

print("Amount of Curved needle malfunctions in our data")
print(curved_needles_malfunction.shape[0])
print()

#bootstraping method 
#test stastitic will be proportion_curved_malfunction - proportion_stright_malfunction
observed_proportion = proportion_curved_malfunction - proportion_stright_malfunction

bootstrap_samples = np.array([])
for x in np.arange(100000):
    bootstrap_sample = np.random.multinomial(total_needles, [0.5,0.5])
    bootstrap_curved = bootstrap_sample[0]/total_needles
    bootstrap_straight=  bootstrap_sample[1]/total_needles
    bootstrap_samples = np.append(bootstrap_samples,bootstrap_curved-bootstrap_straight)


#below is a distribution of the difference in the proportions due to chance if the proportions of curved and straight needles were the same
pd.DataFrame().assign(proportion_differences= bootstrap_samples).plot(kind='hist', bins=20, density=True, ec='w', figsize=(10, 5), title = "Equal Probability Bootstrap Samples");


#With this bootstrap sample we are 95% confidence that the difference in proprotion of a needle being curved given that
#it malfuctioned and the proportion of a needle being stright given that it malfucted is curved should fall in the interval[-0.16901408,  0.16901408]
lower_bound = np.percentile(bootstrap_samples, 2.5)
upper_bound =  np.percentile(bootstrap_samples, 97.5)

confidence_interval = np.array([lower_bound,upper_bound])
print("Our confidence Interval")
print(confidence_interval)
print()

#Since our observed_proportion is 0.6901408450704225 which is outside the interval, we reject the null hypothesis where the 
#two proportions are the same at at 95% confidence level. This means that it is likely that the difference in the proportions
#in our data is likely due to one or more factors other than random chance.
print("Our Observed difference in Proportion")
print(observed_proportion)
print()

#P-value the probability of getting our value equal to or more extreme to our observed propportion due to chance alone
p_value = np.count_nonzero(bootstrap_samples >= observed_proportion)
print("Our p-value")
print(p_value)
#There is an approximately 0% chance that our observed_propotion can be achieved thorugh chance alonge given that the two proportions are the same
```
