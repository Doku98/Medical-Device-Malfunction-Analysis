```python
import pandas as pd
import numpy as np
```

```python
def count_occurrence_in_column(data, column):
    '''
    Given one of the dataframes in the cell below and a column name,
    returns the amount of times specific values in that dataframe occur
    in that column
    
    e.g. count_occurrence_in_column_per_year(data_1999, "event_type") 
    will return how many times different event_types occur in the year 1999
    '''
    data_by_event_type_count = data.groupby(column).count()
    data_by_event_type_count = data_by_event_type_count.assign(
        Count=data_by_event_type_count.get("exemptn_no")).get(["Count"])
    return data_by_event_type_count
```

```python
# Define data types for columns
types = {
    "exemptn_no": str, "report_id": str, "date_of_event": str,
    "dev_prob_cd": str, "model_no": str, "catalog_no": str,
    "impl_avail_for_eval": str, "impl_ret_to_mfr": str
}

# Read CSV files
data_1999 = pd.read_csv("ASR_files/ASR_1999.csv", dtype=types)
data_2000 = pd.read_csv("ASR_files/ASR_2000.csv", dtype=types)
data_2001 = pd.read_csv("ASR_files/ASR_2001.csv", dtype=types)
data_2002 = pd.read_csv("ASR_files/ASR_2002.csv", dtype=types)
data_2003 = pd.read_csv("ASR_files/ASR_2003.csv", dtype=types)
data_2004 = pd.read_csv("ASR_files/ASR_2004.csv", dtype=types)
data_2005 = pd.read_csv("ASR_files/ASR_2005.csv", dtype=types)
data_2006 = pd.read_csv("ASR_files/ASR_2006.csv", dtype=types)
data_2007 = pd.read_csv("ASR_files/ASR_2007.csv", dtype=types)
data_2008 = pd.read_csv("ASR_files/ASR_2008.csv", dtype=types)
data_2009 = pd.read_csv("ASR_files/ASR_2009.csv", dtype=types)
data_2010 = pd.read_csv("ASR_files/ASR_2010.csv", dtype=types)
data_2011 = pd.read_csv("ASR_files/ASR_2011.csv", dtype=types)
data_2012 = pd.read_csv("ASR_files/ASR_2012.csv", dtype=types)
data_2013 = pd.read_csv("ASR_files/ASR_2013.csv", dtype=types)
data_2014 = pd.read_csv("ASR_files/ASR_2014.csv", dtype=types)
data_2015 = pd.read_csv("ASR_files/ASR_2015.csv", dtype=types)
data_2016 = pd.read_csv("ASR_files/ASR_2016.csv", dtype=types)
data_2017 = pd.read_csv("ASR_files/ASR_2017.csv", dtype=types)

# Filter data for 2005 with non-blank brand names
data_2005_no_blank_brand_name = data_2005[
    data_2005["brand_name"].notna()
    ]

# Filter data for 2005 with brand names containing "NEEDLE"
data_2005_needles = data_2005_no_blank_brand_name[
    data_2005_no_blank_brand_name["brand_name"].str.contains("NEEDLE")]


# Count occurrences of needle complications in 2005
amount_of_2005_needles = count_occurrence_in_column(
    data_2005_needles, "brand_name")

# Plot a horizontal bar chart
amount_of_2005_needles.plot(
    kind="barh",
    y="Count",
    fontsize=10,
    title="Amount of Needle Complications in 2005",
    xlabel="Occurrences of Complications",
    ylabel="NEEDLE TYPE"
);
```

```python

# Count occurrences of device complications in 2017 for each manufacturer
manufactures_complications_count_2017 = count_occurrence_in_column(data_2017, "mfr_name")

# Specify the number of manufacturers to show
number_of_manufactures_shown = 15

# Sort and plot the top manufacturers with the most device complications in 2017
manufactures_complications_count_2017.sort_values("Count").tail(number_of_manufactures_shown).plot(
    kind="barh",
    y="Count",
    fontsize=10,
    title="Top 15 Manufacturers With Most Device Complications in 2017",
    xlabel="Occurrences of Complications",
    ylabel="Manufacturer Name"
);
```

```python
# Filter the DataFrame to only contain incidents where serious injuries were due to needles
data_1999_2019_needles_no_empty = data_1999_2019[data_1999_2019[
    "brand_name"].notna()]
data_1999_2019_needles = data_1999_2019_needles_no_empty[
    data_1999_2019_needles_no_empty["brand_name"].str.contains("NEEDLE")]

# Display the filtered data
data_1999_2019_needles

# Report the occurrences for each company when their needle device led to serious injuries
count_occurrence_in_column(data_1999_2019_needles, "brand_name").sort_values("Count").plot(
    kind="barh",
    y="Count",
    fontsize=10,
    title="Needles Which Lead to Most Serious Injuries (IN)",
    xlabel="Occurrences of Complications",
    ylabel="Manufacturer Name"
)

# Further analysis on a specific type of needle (e.g., "CURVED NEEDLE DELIVERY SYSTEM")
data_1999_2019_curved_needles = data_1999_2019_needles[data_1999_2019_needles[
    "brand_name"] == "CURVED NEEDLE DELIVERY SYSTEM"]
```

```python
# The cell below shows how A/B testing 
# (If the difference between needles has the same rate of malfunctions due to random chance or some other factor)

# Null Hypothesis: 
# The probability of a malfunctioning needle being straight is the same as the probability of a malfunctioning needle being curved 
# Alternative Hypothesis:
# The probability of a malfunctioning needle being straight is less than the probability of a malfunctioning needle being curved 


curved_or_straight_needles = data_1999_2019_needles[
    (data_1999_2019_needles.get("brand_name") == "CURVED NEEDLE DELIVERY SYSTEM") |
    (data_1999_2019_needles.get("brand_name") == "STRAIGHT NEEDLE DELIVERY SYSTEM")]

total_needles = curved_or_straight_needles.shape[0]

straight_needles_malfunction = data_1999_2019_needles[
    data_1999_2019_needles.get("brand_name") == "STRAIGHT NEEDLE DELIVERY SYSTEM"]
curved_needles_malfunction = data_1999_2019_needles[
    data_1999_2019_needles.get("brand_name") == "CURVED NEEDLE DELIVERY SYSTEM"]

proportion_straight_malfunction = straight_needles_malfunction.shape[0] / total_needles
proportion_curved_malfunction = curved_needles_malfunction.shape[0] / total_needles

print("Straight of Curved needle malfunctions in our data")
print(straight_needles_malfunction.shape[0])

print("Amount of Curved needle malfunctions in our data")
print(curved_needles_malfunction.shape[0])
print()

# Bootstrapping method 
# Test statistic will be proportion_curved_malfunction - proportion_straight_malfunction
observed_proportion = proportion_curved_malfunction - proportion_straight_malfunction

bootstrap_samples = np.array([])
for x in np.arange(10000):
    bootstrap_sample = np.random.multinomial(total_needles, [0.5, 0.5])
    bootstrap_curved = bootstrap_sample[0] / total_needles
    bootstrap_straight = bootstrap_sample[1] / total_needles
    bootstrap_samples = np.append(bootstrap_samples, bootstrap_curved - bootstrap_straight)

# Below is a distribution of the difference in the proportions
# due to chance if the proportions of curved and straight needles were the same
pd.DataFrame().assign(proportion_differences=bootstrap_samples).plot(
    kind='hist', bins=20, density=True, ec='w', figsize=(10, 5), title="Equal Probability Bootstrap Samples")

# With this bootstrap sample, we are 95% confident that the difference in proportion of a needle being curved given that
# it malfunctioned and the proportion of a needle being straight given that it malfunctioned is curved should fall in the interval[-0.16901408,  0.16901408]
lower_bound = np.percentile(bootstrap_samples, 2.5)
upper_bound = np.percentile(bootstrap_samples, 97.5)

confidence_interval = np.array([lower_bound, upper_bound])
print("Our confidence Interval")
print(confidence_interval)
print()

# Since our observed_proportion is 0.6901408450704225, which is outside the interval, we reject the null hypothesis where the 
# two proportions are the same at at 95% confidence level. This means that it is likely that the difference in the proportions
# in our data is likely due to one or more factors other than random chance.
print("Our Observed difference in Proportion")
print(observed_proportion)
print()

# P-value: the probability of getting our value equal to or more extreme to our observed proportion due to chance alone
p_value = np.count_nonzero(bootstrap_samples >= observed_proportion)
print("Our p-value")
print(p_value)
# There is an approximately 0% chance that our observed_proportion can be achieved through chance alone
# given that the two proportions are the same
```
